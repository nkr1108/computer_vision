## <b>Functions used in PIL</b>

*   open() - It is used to read an image from the given path.
*   load() - It is used to extract the pixel map from the image.

*   display() - It is used to display an image.
*   size - It returns the width and height of an image.

*   getpixel() - It loops over the pixel values and returns the values of the three channels.
*   save() - It is used to save an image.

*   convert() - It is used to convert an image into the required format. For example, converting an RGB image to HSV.
*   getchannel() - It is used to get each individual channel from the image.

**Step 1 - Import Libraries**

```python
# Importing the Libraries
import PIL
from PIL import Image
import matplotlib.pyplot as plt
```

**Step 2 - Open Image**

```python
# Loading the path of the image
image_path = '/content/drive/MyDrive/Colab Notebooks/ch_8_computer_vision/test_image.png'
#image_path = 'test_image.png'

# Import an image
image = Image.open(image_path)
#image = Image.open('test_image.png')
```

**Step 3 - load - extract the pixel map from the image**

```python
# Extracting pixel map:
pixel_matrix = image.load()
```

**Step 6 - getpixel() - returns values of 3 channels - Conversion of RGB to Grayscale**


<b> The grayscale image is obtained by finding the average of the three channels of an RGB image.</b>
*   The Mathematical Formula to convert an RGB image to grayscale is as shown below:

<b>
<center><h1>
$\frac{(R+G+B)}{3}$



```python
for i in range(width):
    for j in range(height):
        
        # getting the RGB pixel values.
        r, g, b,= image.getpixel((i, j))
          
        # Apply formula of grayscale:
        
        grayscale = (r+g+b)/3  # grayscale normal average

        # setting the pixel value.
        pixel_matrix[i, j] = (int(grayscale), int(grayscale), int(grayscale))
  
```

```python
display(image)
# pixels are from left to right - width, and top to bottom - height
```


*   This common formula for obtaining a grayscale image is theoretically correct, but an improved formula has been obtained by researchers from the weighted average of the pixel intensities of red, blue, and green colors which can be shown as:

<b>
<center><h1>
${(0.299R+0.587G+0.114B)}$





```python
for i in range(width):
    for j in range(height):
        
        # getting the RGB pixel values.
        r, g, b,= image.getpixel((i, j))
          
        # Apply formula of weighted wavelengths to obtain grayscale:

        grayscale = (0.299*r + 0.587*g + 0.114*b)   # grayscale weighted average

        # setting the pixel value.
        pixel_matrix[i, j] = (int(grayscale), int(grayscale), int(grayscale))
  
```

**Step 7 - saving the image object into png file**

```python
# Saving the final output as "grayscale.png":
image.save("grayscale.png", format="png")
```
